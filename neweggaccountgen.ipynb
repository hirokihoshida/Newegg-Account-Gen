```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import WebDriverException, TimeoutException, NoSuchElementException

import requests
import pandas as pd
import numpy as np
import time
import math

import urllib.parse as urlparse
from urllib.parse import parse_qs
```

```python
information = pd.read_csv('information.csv')
```

```python
output = pd.DataFrame()
for index, info in information.iterrows():
    print(info)
    
    #disable css/img
    firefox_profile = webdriver.FirefoxProfile()
    firefox_profile.set_preference('permissions.default.image', 2)
    firefox_profile.set_preference('permissions.default.stylesheet', 2)
    
    #if proxy in info, use proxy
    if info['proxy'] != info['proxy']:
        driver = webdriver.Firefox(firefox_profile=firefox_profile)
    else:
        #if auth proxy
        if info['proxy'].count(":") != 1:
            print("Invalid proxy (Auth) \n")
            continue
        #set proxy
        firefox_capabilities = webdriver.DesiredCapabilities.FIREFOX
        firefox_capabilities['proxy'] = {
            "proxyType": "MANUAL",
            "httpProxy": info['proxy'],
            "ftpProxy": info['proxy'],
            "sslProxy": info['proxy']
        }
        driver = webdriver.Firefox(capabilities=firefox_capabilities, firefox_profile=firefox_profile)
        
    #go to site
    driver.set_page_load_timeout(20)
    try:
        driver.get("https://www.newegg.com/")

        #if country menu shows
        if len(driver.find_elements_by_xpath("//button[text() = 'Stay at United States']")) > 0:
            print("Non US Proxy \n")
            driver.close()
            continue

#         #if ip banned
#         if len(driver.find_elements_by_xpath("//p[contains(text(),'automated program to visit')]")) > 0:
#             print("404 Forbidden Error \n")
#             driver.close()
#             continue  

        try:
            #click account login button
            login = driver.find_element_by_xpath("//a[@href='https://secure.newegg.com/NewMyAccount/AccountLogin.aspx?nextpage=https%3A%2F%2Fwww.newegg.com%2F']")
            login.click()
            #click account signup button
            login = driver.find_element_by_xpath("//strong[text() = 'Sign Up']/parent::*")
            login.click()
        except NoSuchElementException:
            print("404 Forbidden Error \n")
            driver.close()
            continue 

        #enter information
        inputField = driver.find_element_by_xpath("//label[text()='First Name']/following-sibling::input")
        inputField.send_keys(info['fname'])
        time.sleep(0.5)
        inputField = driver.find_element_by_xpath("//label[text()='Last Name']/following-sibling::input")
        inputField.send_keys(info['lname'])
        time.sleep(0.5)
        inputField = driver.find_element_by_xpath("//label[text()='Email Address']/following-sibling::input")
        inputField.send_keys(info['email'])
        time.sleep(0.5)
        inputField = driver.find_element_by_xpath("//label[text()='Mobile Phone Number (optional)']/following-sibling::input")
        inputField.send_keys(info['phone'])
        time.sleep(0.5)
        inputField = driver.find_element_by_xpath("//label[text()='Password']/following-sibling::input")
        inputField.send_keys(info['password'])
        time.sleep(0.5)
        #if newsletter = false, uncheck newsletter box
        if info['newsletter'] == False:
            driver.execute_script("document.querySelectorAll('input[type=checkbox]')[0].click()")
            time.sleep(0.5)
        #submit
        driver.execute_script("document.querySelectorAll('button[type=submit]')[0].click()")

        #wait until account already made error or home page redirect
        try:
            WebDriverWait(driver, 10).until(lambda driver: driver.find_elements_by_xpath("//a[contains(text(),'Components')]") or driver.find_elements_by_xpath("//p[contains(text(),'already registered')]"))
        except:
            print("Timeout Error \n")
            driver.close()
            continue
        #if account already made
        if len(driver.find_elements_by_xpath("//p[contains(text(),'already registered')]")) > 0:
                print("Account already made \n")
                driver.close()
                continue
        #if successful
        output = output.append(info)
        driver.close()
        print("Account Created \n")
        output.to_csv('output.csv',index=False)
    except (WebDriverException, TimeoutException) as e:
        print("Bad Proxy \n")
        driver.close()
        continue
```
